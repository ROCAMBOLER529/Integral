```python
# Integral Triple
# \int_{D}^{} f(x, y, z) dydzdx = \int_{D}^{} xyz dydzdx

def f(x, y, z):
    return x*y*z;

display(Math("\int_{D}^{} xyz dydzdx"));
display(Math("\int_{0}^{3} \int_{1}^{x-3} \int_{0}^{z² - 4} (xyz) dydzdx"));

i = integrate.tplquad(f, 0, 3, 1, lambda x: x-3, 0, lambda x, y: x**2 - 4);
print("Resultado de la intergal triple: ", i[0]);

fig = plt.figure(figsize=(6, 4));
ax = plt.axes(projection = '3d');
```

```python
# Plano Tangente a una Superficie
# f(x, y, z) = 5x² + 5y² - 25z² = -125

#Superficie

x0 = int(input("Ingrese un valor para x0: "));
y0 = int(input("Ingrese un valor para y0: "));
z0 = math.sqrt((5*x0**2 + 5*y0**2 + 125) / 25);

dx = 10*x0;
dy = 10*y0;
dz = -50*z0;

a = np.linspace(-50, 50, 50);
b = np.linspace(-50, 50, 20);

def superficie(x, y):
    return np.sqrt((5*x**2 + 5*y**2 + 125) / 25);

x, y = np.meshgrid(a, b);
z = superficie(x, y);

fig = plt.figure(figsize=(6, 4));
ax = plt.axes(projection = '3d');

ax.plot_surface(x, y, z, color="r");

# Plano Tangente
def tangente(x, y):
    return (dx*(x - x0) + dy*(y - y0))/(dz + z0);

x, y = np.meshgrid(a, b);
z = tangente(x, y);
ax.plot_surface(x, y, z, color="b");

ax.scatter3D(x0, y0, z0, s = 80, edgecolor = "y", facecolor = "gold");

ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');
ax.set_xlim(-250, 250);
ax.set_ylim(-250, 250);
ax.set_zlim(-250, 250);
ax.zaxis.set_major_locator(LinearLocator(10));

plt.show();
```
